```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# R. Stull,  23 Feb 2016
# ES = equation number in BAMS supplementary material by Stein et al (Dec 2015) on Hysplit model
# http://journals.ametsoc.org/doi/suppl/10.1175/BAMS-D-14-00110.1/suppl_file/10.1175_bams-d-14-00110.2.pdf

# Givens:
# general
tlhoriz = 10800.    # lagrangian timescale for horizontal turbulence (s)

g_by_tv = 0.0333    # g/Tv approx constant  (m s^-2 K^-1)
rho_cp = 1231.      # air density times specific heat at const press, approx as constant. (W/m2)/(K m/s)

# stable and neutral pbl
ustar_n = 0.2          # friction velocity (m/s) for neutral PBL with moderate mean wind
zi_n = 500.            # pbl depth in neutral and stable
tlvert_s = 5.          # lagrangian time scale for vertical motions for stable pbl (s)
tlvert_n = 200.        # lagrangian time scale for vertical motions for unstable pbl (s)

# unstable pbl
Hsfc = 280.             # surface heat flux (W/m2)
FHsfc = Hsfc / rho_cp   # kinematic heat flux (K m/s) at the surface
ustar_u = 0.05          # friction velocity (m/s) for convective PBL with light winds
zi_u = 1000.            # mixed-layer depth for unstable pbl (m)
tlvert_u = 200.         # lagrangian time scale for vertical motions for unstable pbl (s)

# Hint 1: for unstable pbl, use w* = Deardorff velocity from Stull Practical Meteor, eq (18.19a)
# Hint 2: kinematic heat flux FHsfc = (traditional heat flux) / rho_cp . 
# Hint 3: for the Obukhov length (L), use  z/L = -0.4 * (z/zi) * (w*/u*)^3
# Hint 4: the "surface layer" is generally the bottom 10% of the whole boundary layer
```

### (1) stable/neutral pbl on one graph, using eqs (ES9 - ES11)

The equations referenced are:

$$
\sigma_w^2 = 1.7u_*^2(1-z/z_i)^\frac{3}{2}\tag{ES9}
$$

$$
\sigma_u^2 = 4.0u_*^2(1-z/z_i)^\frac{3}{2}\tag{ES10}
$$

$$
\sigma_v^2 = 5.0u_*^2(1-z/z_i)^\frac{3}{2}\tag{ES11}
$$



```python
# initialize z
z = np.linspace(0, 1, 100)

# get sigma
sigma_w2 = 1.7 * ustar_n ** 2 * (1 - z / np.max(z)) ** (3 / 2)  # ES9
sigma_u2 = 4.0 * ustar_n ** 2 * (1 - z / np.max(z)) ** (3 / 2)  # ES10
sigma_v2 = 5.0 * ustar_n ** 2 * (1 - z / np.max(z)) ** (3 / 2)  # ES11

# plot it
plt.plot(sigma_w2, z, label="$\sigma_w^2$")
plt.plot(sigma_u2, z, label="$\sigma_u^2$")
plt.plot(sigma_v2, z, label="$\sigma_v^2$")

plt.ylabel("$z/z_i$")
plt.xlabel("velocity variance ($m^2/s^2$)")
plt.title("Turbulent Velocity Variance in the Stable/Neutral PBL")

plt.legend()
```

**Discussion**

*In the absence of convection, we expect velocity variance to be generated mechanically from wind shear. Shear ($\frac{dv_i}{dx_i}$) is typically at a maximum near the surface, and fades as we approach the free atmosphere (constant wind). This parameterization has the smallest $\sigma$ value in z, as stable atmospheres supress vertical movement. We also expect more dispersion in the crosswind direction than in the downwind, consistent with this parameterization.*

### (2) unstable pbl on another graph, using eqs. (ES15 - ES16)

Implement the following equations:

$$
\sigma_w^2 = w_*^2 (z/z_i)^\frac{2}{3} (1 - z/z_i)^\frac{2}{3} \left[1 + 0.5(H_{PBL}/H_{SFC})^\frac{2}{3}\right] \tag{ES15}
$$

$$
\sigma_u^2 = \sigma_v^2 = 0.36w_*^2 \tag{ES16}
$$

$$
w_* = \left[\frac{|g|}{T_v}\cdot z_i \cdot F_{Hsfc}\right]^\frac{1}{3}\tag{Stull 18.19}
$$

```python
# Hint 1: for unstable pbl, use w* = Deardorff velocity from Stull Practical Meteor, eq (18.19a)
w_star = (g_by_tv * np.max(z) * FHsfc) ** (1 / 3)

# As per supplement Hpbl / Hsfc is ~ constant
Hpbl_by_Hsfc = 0.2

# ES15
sigma_w2 = (
    w_star ** 2
    * (z / np.max(z)) ** (2 / 3)
    * (1 - z / np.max(z)) ** (2 / 3)
    * (1 + 0.5 * (Hpbl_by_Hsfc) ** (2 / 3))
)

# ES16
sigma_u2 = np.ones(len(z)) * 0.36 * w_star ** 2 
sigma_v2 = np.ones(len(z)) * 0.36 * w_star ** 2

# plot it
plt.plot(sigma_w2, z, label="$\sigma_w^2$")
plt.plot(sigma_u2, z, label="$\sigma_u^2$, $\sigma_v^2$")
#plt.plot(sigma_v2, z, label="$\sigma_v^2$")

plt.ylabel("$z/z_i$")
plt.xlabel("velocity variance ($m^2/s^2$)")
plt.title("Turbulent Velocity Variance in the Convective PBL")

plt.legend();
```

**Discussion**

*For the free convective case, we expect the largest generation of buoyant turbulence production to occur some distance above the surface, where heat flux is large and large eddies are able to form. Near $z_i$, we observe a decrease in $\sigma_w$ where turbulence is being consumed by eating away at the BL inversion*

*Horizontal velocity variance is captured by a very simple parameterization -- taken to be constant*

### (3) unstable surface layer in the previous graph, using eq. (ES17)

Implement ES17:

$$
\sigma_w^2 = 0.74u_*^2(1-3z/L)^\frac{2}{3}\tag{ES17}
$$

From the hint:

$$
z/L = -0.4 \cdot (z/z_i) \cdot (w_*/u_*)^3
$$

```python
# construct the surface layer (10% of boundary layer)
z_sfc = np.linspace(0, np.max(z) * 0.1, 100)

# get z/L
z_by_L = -0.4 * (z_sfc / np.max(z)) * (w_star / ustar_u) ** 3

# ES17
sigma_w2_new = 0.74 * ustar_u ** 2 * (1 - 3 * z_by_L) ** (2 / 3)

# plot it
plt.plot(sigma_w2_new, z_sfc, label="$\sigma_w^2$")
#plt.plot(sigma_u2, z, label="$\sigma_u^2$")
#plt.plot(sigma_w2, z, label="$\sigma_w^2$")

#plt.plot(sigma_v2, z, label="$\sigma_v^2$")

plt.ylabel("$z/z_i$")
plt.xlabel("velocity variance ($m^2/s^2$)")
plt.title("Turbulent Velocity Variance in the Unstable Surface Layer")
plt.plot

plt.legend();
```

**Discussion**

*Turbulence near the surface should follow a "law of the wall" -- where the veloctiy of a fluid drops to zero near a rigid boundary. As we move away from the surface, larger eddies are able to form uninhibited by the surface, and so $\sigma$ increases.*
