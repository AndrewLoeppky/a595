## (A) 
Plot the dimensionless Zcl vs. X curves (analogous to Stull Fig. 19.7), with separate curves
for the two cases (morning, midday).

Hint: Increment X between 0 and 6 (using very small delta-X at small X,
and larger deltz-X as X gets larger),
and apply to eq. (19.32).  Use 0 ≤ Z ≤ 1 for the range of dimensionless heights.


$$
Z_{CL} \approx 0.5 + \frac{0.5}{1 + 0.5\cdot X^2}\cdot cos\left[2\pi\frac{X}{\lambda} + cos^{-1}(2Z_s - 1)\right]\tag{19.32}
$$


```python
### Create Morning and Midday Cases ###

morning = {"g_by_tv":0.0333,          # g/Tv approx. constant  (m s^-2 K^-1)
           "m":2,                     # wind speed (m/s)  in (morning , mid-day)
           "zi":100.,                 # mixed layer depth (m) 
           "FH":0.05,                 # effective surface kinematic heat flux (K m/s)
           "zs":75.,                  # source height of pollutants after plume rise (m)
           "q":3e5,                  # emission rate (mg/s)
           "xmax_d":6.,               # dimensionless max distance X downwind, for Deardorff's calculations
           "ymax_d":5.}               # dimensionless max crosswind range, for Y within +- ymax_d.
           
midday = {"g_by_tv":0.0333,           # g/Tv approx. constant  (m s^-2 K^-1)
           "m":5,                     # wind speed (m/s)  in (morning , mid-day)
           "zi":1000.,                # mixed layer depth (m) 
           "FH":0.2,                  # effective surface kinematic heat flux (K m/s)
           "zs":75.,                  # source height of pollutants after plume rise (m)
           "q":3e5,                  # emission rate (mg/s)
           "xmax_d":6.,               # dimensionless max distance X downwind, for Deardorff's calculations
           "ymax_d":5.}               # dimensionless max crosswind range, for Y within +- ymax_d.
```

```python
def get_centerline(g_by_tv, m, zi, FH, zs, q, xmax_d, ymax_d):
    '''
    Computes plume centerline as per Stull 19.32
    '''
    
    # initialize coordinate system, more points close to the plume
    x = np.logspace(0,6,1000)
    #x = np.linspace(0.1,20000,1000)
    
    # compute non-dimensional quantities
    wstar = (g_by_tv * zi * FH) ** (1 / 3)
    Zs = zs / zi
    X = x * wstar / (zi * m)
    lmda = 4
    
    # Stull 19.32
    Zcl = 0.5 + (0.5 / (1 + 0.5 * X ** 2)) * np.cos(2 * np.pi * X / lmda + np.arccos(2 * Zs - 1))
    return X, Zcl

# do the plot
Xmorn, Zclmorn = get_centerline(**morning)
Xmid, Zclmid = get_centerline(**midday)
plt.plot(Xmorn, Zclmorn, color="k", linestyle="--", alpha=0.7, label="Morning")
plt.plot(Xmid, Zclmid, color="k", linestyle=":", alpha=0.9, label="Midday")
plt.xlabel("$X=xw_s / z_im$")
plt.ylabel("$Z_{CL}=z_{CL}/z_i$")
plt.title("Dimensionless Plume Centerline, Morning and Midday")
plt.legend()
plt.xlim(0, 6)
plt.ylim(0,1);
```

## (B) 
Find and plot the dimensionless cross-wind integrated concentration contours (similar to Stull Fig. 19.8), but for dimensionless release heights corresponding to the two cases (morning, mid-day), for dimensionless ranges:  0 ≤ X ≤ 6  and 0 ≤ Z ≤ 1 .

$$
C_y' = exp\left[-0.5\cdot\left(\frac{Z - Z_{CL}}{\sigma_{zdc}'}\right)^2\right]\tag{19.33}
$$

$$
\sigma_{zdc}' = a\cdot X = 0.25\cdot X\tag{19.34}
$$

For $K$ evenly spaced vertical levels:

$$
\overline{C_y'} = \frac{1}{K}\sum_{k=1}^K C_y'\tag{19.35}
$$

Finally, calculate dimensionless concentration via

$$
C = \frac{C_y}{(2\pi)^{1/2}0.5\cdot X}exp\left[-0.5\cdot\left(\frac{Y}{0.5\cdot X}\right)^2\right]\tag{19.37*}
$$


```python
def get_CIC(g_by_tv, m, zi, FH, zs, q, xmax_d, ymax_d):
    '''
    Computes the crosswind integrated concentration (CIC) as per stull 19.33 - 19.37
    '''
    X, Zcl = get_centerline(g_by_tv, m, zi, FH, zs, q, xmax_d, ymax_d)
    K = 50  # number of vertical layers
    Z = np.linspace(0, 1, K)
    
    # initialize a concentration grid
    XZ = np.zeros((K,len(X)))
    for i in range(K):
        XZ[i,:] = i / K
    
    # calculate Cy' (19.33)
    Cyprime = np.exp(-0.5 * ((XZ - Zcl) / (0.25 * X)) ** 2)
    
    # calculate \overbar{Cy'} and Cy (19.35, 19.36)
    Cypb = np.zeros_like(Cyprime)
    mean1d = np.mean(Cyprime, axis=0)
    for i, cy in enumerate(mean1d):
        Cypb[:,i] = cy
    
    # final product
    Cy = Cyprime / Cypb
    return X, Z, Cy

# use the function for both cases
X1, Z1, Cymorn = get_CIC(**morning)
X2, Z2, Cymid = get_CIC(**midday)

# plot results
fig, (ax1, ax2) = plt.subplots(2, figsize=(15,8), sharex=True)
#fig.tight_layout()
fig.suptitle("Crosswind Integrated Dimensionless Concentration \n\n \
Top: Morning        Bottom: Midday",fontsize=15)

p1 = ax1.contour(X1, Z1, Cymorn, levels=np.linspace(0.1,5,30),cmap="gray", linewidths=1)
p2 = ax2.contour(X2, Z2, Cymid, levels=np.linspace(0.1,5,30),cmap="gray", linewidths=1)
ax1.clabel(p1, p1.levels, inline=False, fontsize=10)
ax2.clabel(p2, p2.levels, inline=False, fontsize=10)

ax1.set_xlim(0,6)
ax1.set_ylim(0,1)
ax2.set_xlim(0,6)
ax2.set_ylim(0,1)

ax2.set_xlabel("$X=xw_s / z_im$")
ax1.set_ylabel("$Z=z/z_i$")
ax2.set_ylabel("$Z=z/z_i$");
```

## (D) 

Create contour plots of the dimensionless concentration C at the earth's surface (Z = 0), for a surface map spanning the following region: dimensionless coordinates of 0 ≤ X ≤ 6 and -3 ≤ Y ≤ 3 , using a separate plot for each case (morning, midday). Plot the concentration isopleths for dimensionless C values of  levels = c( seq(0.05, 0.35, 0.05) , seq(0.4, 2.0, 0.2), seq(3, 10, 1) )
 
*Note: this has nothing to do with exercise (D), but it can be built on your results for exercise (B).*

$$
C = \frac{C_y}{(2\pi)^{1/2}\cdot 0.5\cdot X}exp\left[-0.5\cdot\left(\frac{Y}{0.5\cdot X}\right)^2\right]\tag{19.37*}
$$

```python
def get_surfconc(g_by_tv, m, zi, FH, zs, q, xmax_d, ymax_d):
    '''
    computes the dimensionless surface concentration on 0 ≤ X ≤ 6 and -3 ≤ Y ≤ 3
    as per Stull 19.37
    '''
    X, Z, Cy = get_CIC(g_by_tv, m, zi, FH, zs, q, xmax_d, ymax_d)
    Cy0 = Cy[0] # get Cy at ground level
    J = 51 # Number of levels in Y direction
    Y = np.linspace(-ymax_d,ymax_d, J)
    
    XY = np.zeros((J, len(X))) # create the XY plane
    for i, y in enumerate(Y):
        XY[i] = y # assign Y values to XY
    
    # stull 19.37
    C = Cy0 / (np.sqrt(2 * np.pi) * 0.5 * X) * np.exp(-0.5 * (XY / (0.5 * X)) ** 2)
    return X, Y, C

# use the function for both cases
X1, Y1, C_morn = get_surfconc(**morning)
X2, Y2, C_mid = get_surfconc(**midday)

# create contour levels as spec'd in question
mylevels = np.concatenate((np.arange(0.05, 0.40, 0.05),
                           np.arange(0.4, 2.2, 0.2),
                           np.arange(3, 11, 1)),
                           axis=None)

# do the plot
fig, (ax1, ax2) = plt.subplots(2, figsize=(15,8), sharex=True)
fig.suptitle("Dimensionless Concentration At Ground\n\n \
Top: Morning        Bottom: Midday",fontsize=15)

p1 = ax1.contour(X1, Y1, C_morn, levels=mylevels,cmap="gray", linewidths=1)
p2 = ax2.contour(X2, Y2, C_mid, levels=mylevels,cmap="gray", linewidths=1)
ax1.clabel(p1, p1.levels, inline=False, fontsize=10)
ax2.clabel(p2, p2.levels, inline=False, fontsize=10)

ax1.set_xlim(0,6)
ax1.set_ylim(-3,3)
ax2.set_xlim(0,6)
ax2.set_ylim(-3,3)
ax2.set_xlabel("$X=xw_s / z_im$")
ax1.set_ylabel("$Y=y/z_i$")
ax2.set_ylabel("$Y=y/z_i$");
```

## (E) 

Same as D, but plot contours for actual concentrations c (mg/m3) vs. actual distances x (m) and y (m). Hints: Your plot domain could be  for 0 ≤ x ≤ 3,000 m , and  -1500 ≤ y ≤ 1500 m . You can draw the concentration contours for the following values: c (mg/m3) = 0.05, 0.1, 0.2,  0.3, 0.5,  1.0 , 2.0 , 3.0 , 5.0 , 10.0 , 15.0, 20.0 , 30.0

$$
C = \frac{c\cdot z_i^2\cdot M}{Q} \rightarrow c = \frac{C\cdot Q}{z_i^2\cdot M}   \tag{19.23}
$$

$$
C_y = \frac{c_y\cdot z_i\cdot M}{Q} \rightarrow c_y = \frac{C_y\cdot Q}{z_i \cdot M}\tag{19.24}
$$

$$
X = \frac{x\cdot w_*}{z_i \cdot M} \rightarrow x = \frac{Xz_iM}{w_*}\tag{19.25}
$$

$$
Y = \frac{y}{z_i}\rightarrow y = Yz_i\tag{19.26}
$$

$$
Z = \frac{z}{z_i}\rightarrow z = Zz_i\tag{19.27}
$$

$$
Z_{CL} = \frac{z_{CL}}{z_i}\rightarrow z_{CL} = Z_{CL}z_i\tag{19.28}
$$

```python
def dimensionalize_plume(g_by_tv, m, zi, FH, zs, q, xmax_d, ymax_d):
    '''
    takes in a non dimensional plume and converts to SI units
    '''
    # create ground level case
    X, Y, C = get_surfconc(g_by_tv, m, zi, FH, zs, q, xmax_d, ymax_d)
    wstar = (g_by_tv * zi * FH) ** (1 / 3)
    
    # do the conversions and return result
    x = X * zi * m / wstar  # 19.25
    y = Y * zi  # 19.26
    c = C * q / (zi ** 2 * m)  # 19.23
    
    return x, y, c

# use the function for both cases
x1, y1, c_morn = dimensionalize_plume(**morning)
x2, y2, c_mid = dimensionalize_plume(**midday)

# create contour levels as spec'd in question
mylevels = np.array([0.05, 0.1, 0.2, 0.3, 0.5, 1.0 , 2.0 , 3.0 , 5.0 , 10.0 , 15.0, 20.0 , 30.0])

# do the plot
fig, (ax1, ax2) = plt.subplots(2, figsize=(15,8), sharex=True)
fig.suptitle("Dimensionless Concentration At Ground\n\n \
Top: Morning        Bottom: Midday",fontsize=15)

p1 = ax1.contour(x1, y1, c_morn, levels=mylevels, cmap="gray", linewidths=1)
p2 = ax2.contour(x2, y2, c_mid, levels=mylevels, cmap="gray", linewidths=1)
ax1.clabel(p1, p1.levels, inline=False, fontsize=10)
ax2.clabel(p2, p2.levels, inline=False, fontsize=10)

ax1.set_xlim(0,3000)
ax1.set_ylim(-1500,1500)
ax2.set_xlim(0,3000)
ax2.set_ylim(-1500,1500)
ax2.set_xlabel("Downwind Distance (m)")
ax1.set_ylabel("Crosswind Distance (m)")
ax2.set_ylabel("Crosswind Distance (m)");
```

## (G) 

Discuss:
    
   (1) Why does C or c decrease seemingly monotonically with increasing distance X or x
 downwind of the first maximum, even though
 the crosswind-integrated concentrations Cy of exercises (B) or (C) seem to show
 a secondary relative minimum at distances of roughly X = 1.5 to 2.5 (dimensionless)
 with a secondary relative maximum beyond that ?
 
  > *Because the plume spreads out in the crosswind direction! Integrating over Y picks up
  "high" concentrations over a larger and larger domain as we move downwind. The lateral (Y)
  spreading of the plume happens faster than the secondary oscillation (in Z), so a receptor
  at a single point observes a monotonic decrease with distance downwind (X)*
   
  (2) Why do the morning and mid-day cases give the max concentrations at roughly the same
 real x (m) location downwind, even though the cross-wind integrated
 concentrations for the mid-day case
 bring the high concentrations down to the surface much sooner than for the morning case?
 
   >*Non-dimensional distance X is a function of windspeed, mixed layer depth and deardorff 
   velocity, all of which change from morning to midday. The non-dimensional max concentrations 
   are in different (nondimensional) locations, but the contributing factors more-or-less cancel
   in this case, so the max concentrations end up very close to one another in physical space*
   
  (3) For exercise (E) for the morning case, if you plotted the domains as I
 specified in the exercises, then why do the contour lines seem to be cut off
  before extending their full width from the centerline?
  
   >*The spatial coordinates are larger than the non-dimensional ones, so converting from Y to y 
    runs off the end of the domain of Y. Increasing `ymax_d` fixes this* 
   
  (4) Given your contour plots from exercise (E), what is the primary advantage of using this
 convective dispersion method (i.e., Deardorff, as approximated by Stull) over using a traditional
 Gaussian plume? 
 
   > *This model is a higher order approximation to real meteorology, where the plume centerline 
   demonstrates realistic behavior (general downward motion due to widespread, slow sink and infrequent, sharp lift), and spread occurs around it. This method is more faithful
   to reality than the somewhat crude Pasquill-Guilford sigma values for a plume spreading around a flat centerline*
